```python
# Cell 1: Load Sample Data
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
df.head()
```

```python
# Cell 2: Train/Test Split and Random Forest Training for flower classification
X = df.drop("species", axis=1)
y = df["species"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Cell 3: Feature Importance
importances = model.feature_importances_
plt.bar(X.columns, importances)
plt.title("Feature Importances")
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.show()
```

```python
# Cell 4: Exponential Decay Fit
import numpy as np
from scipy.optimize import curve_fit

# Simulated decay data
time = np.linspace(0, 10, 100)
true_half_life = 3  # true half-life (T)
decay = np.exp(-np.log(2) * time / true_half_life)
noise = 0.05 * np.random.randn(len(decay))
observed = decay + noise

# Fit model: N(t) = exp(-ln(2) * t / T)
def decay_model(t, T):
    return np.exp(-np.log(2) * t / T)

params, _ = curve_fit(decay_model, time, observed, p0=[2])
fitted_half_life = params[0]

# Plot result
plt.plot(time, observed, label='Observed')
plt.plot(time, decay_model(time, *params), label=f'Fitted (T={fitted_half_life:.2f})')
plt.title('Radioactive Decay Fit')
plt.xlabel('Time')
plt.ylabel('Activity')
plt.legend()
plt.show()
```
